### An example showing the plot_pca_2d_projection method used by a scikit-learn classifier

In this example, we'll be plotting a `pca_2d_projection` to describe the classifier model performance
using the **breast_cancer**, **iris** or **digits** dataset from scikit-learn.
- `LogisticRegression`
- `RandomForestClassifier`

Here, we'll be using the `scikitplot.decomposition.plot_pca_2d_projection` method.

```python
from sklearn.datasets import (
    make_classification,
    load_breast_cancer as data_2_classes,
    load_iris as data_3_classes,
    load_digits as data_10_classes,
)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import numpy as np; np.random.seed(0)
# importing pylab or pyplot
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 5)

# Import scikit-plot
import scikitplot as sp
sp.__version__
```

```python
# Load the data
X, y = data_10_classes(return_X_y=True, as_frame=True)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.5, random_state=0)

# Create an instance of the PCA
pca = PCA(random_state=0).fit(X_train)
```

```python
# Plot!
ax = sp.decomposition.plot_pca_2d_projection(
    pca, X_train, y_train, biplot=0, feature_labels=X.columns.tolist()
);
```

## LinearDiscriminantAnalysis

```python
# Load the data
X, y = data_10_classes(return_X_y=True, as_frame=True)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.5, random_state=0)

# Create an instance of the LinearDiscriminantAnalysis
clf = LinearDiscriminantAnalysis().fit(X_train, y_train)
```

```python
# Plot!
ax = sp.decomposition.plot_pca_2d_projection(
    clf, X_train, y_train, biplot=1, feature_labels=X.columns.tolist()
);
```

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# 1. PCA Example
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X)

# Access PCA components (eigenvectors)
print("PCA Components:")
print(pca.components_)  # Shape: (n_components, n_features)

# 2. LDA Example
lda = LDA(n_components=2)  # Reduce to 2 dimensions (maximum for LDA)
X_lda = lda.fit_transform(X, y)

# Access LDA scalings (similar to components in PCA)
print("\nLDA Scalings (Eigenvectors for transformation):")
print(lda.scalings_.T)  # Shape: (n_features, n_components)
```
