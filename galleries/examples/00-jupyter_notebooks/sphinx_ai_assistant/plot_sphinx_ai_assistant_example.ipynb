```python
import os
import tempfile
from pathlib import Path

# Create a temporary directory that gets deleted when the 'with' block ends
with tempfile.TemporaryDirectory() as tmp_dir:
    tmp_path = Path(tmp_dir)
    
    # 1. Define the subdirectory
    site = tmp_path / "plain"
    site.mkdir()
    
    # 2. Write the file
    page = site / "page.html"
    page.write_text(
        '<html><body><main>Plain content</main></body></html>', 
        encoding="utf-8"
    )
    
    print(f"File created at: {page}")
    print(f"Content: {page.read_text()}")

    from scikitplot._externals._sphinx_ext._sphinx_ai_assistant import (
        process_html_directory,
    )
    stats = process_html_directory(
        site,
        theme_preset="plain_html",
        generate_llms=True,
        base_url="https://example.com",
    )
    print(stats)  # {"generated": 42, "skipped": 3, "errors": 0}
    print(os.listdir(site))
```

```python
    from pathlib import Path
    def _make_corpus(root: Path) -> Path:
        """Minimal 3-page corpus with internal links and headings."""
        (root / "index.md").write_text(
            "# Home\n\nWelcome.\n\n[Guide](guide.md)\n",
            encoding="utf-8",
        )
        (root / "guide.md").write_text(
            "# Guide\n\n## Installation\n\nSee [API](api/module.md).\n",
            encoding="utf-8",
        )
        sub = root / "api"
        sub.mkdir()
        (sub / "module.md").write_text(
            "# Module Reference\n\nNo links here.\n",
            encoding="utf-8",
        )
        return root
```

```python
import tempfile
import os
# Create a temporary directory that gets deleted when the 'with' block ends
with tempfile.TemporaryDirectory() as tmp_dir:
    tmp_path = Path(tmp_dir)
    _make_corpus(tmp_path)
    print(os.listdir(tmp_path))
    print(ai.plot_corpus_knowledge(tmp_path, include_links=False))
    g = ai.plot_corpus_knowledge(tmp_path, include_links=False)
    print(g["stats"]["isolated_pages"])
```
