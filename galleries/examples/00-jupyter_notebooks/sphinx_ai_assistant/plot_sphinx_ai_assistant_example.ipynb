```python
import os
import tempfile
from pathlib import Path

# Create a temporary directory that gets deleted when the 'with' block ends
with tempfile.TemporaryDirectory() as tmp_dir:
    tmp_path = Path(tmp_dir)
    
    # 1. Define the subdirectory
    site = tmp_path / "plain"
    site.mkdir()
    
    # 2. Write the file
    page = site / "page.html"
    page.write_text(
        '<html><body><main>Plain content</main></body></html>', 
        encoding="utf-8"
    )
    
    print(f"File created at: {page}")
    print(f"Content: {page.read_text()}")

    from scikitplot._externals._sphinx_ext._sphinx_ai_assistant import (
        process_html_directory,
    )
    stats = process_html_directory(
        site,
        theme_preset="plain_html",
        generate_llms=True,
        base_url="https://example.com",
    )
    print(stats)  # {"generated": 42, "skipped": 3, "errors": 0}
    print(os.listdir(site))
```
