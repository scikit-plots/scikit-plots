```python
import numpy as np
from openTSNE import TSNE
from openTSNE.affinity import PerplexityBasedNN
from openTSNE.nearest_neighbors import PrecomputedNeighbors

def knn_from_annoy(idx, k, search_k=-1):
    n = idx.get_n_items()
    nbrs = np.empty((n, k), dtype=np.int32)
    dists = np.empty((n, k), dtype=np.float32)

    for i in range(n):
        ids, ds = idx.get_nns_by_item(i, k + 1, search_k=search_k, include_distances=True)

        # drop self neighbor if present
        if len(ids) > 0 and ids[0] == i:
            ids, ds = ids[1:], ds[1:]

        ids = ids[:k]
        ds = ds[:k]

        if len(ids) != k:
            raise ValueError(f"Item {i}: expected {k} neighbors, got {len(ids)}")

        nbrs[i] = ids
        dists[i] = ds

    return nbrs, dists

perplexity = 30
k = int(3 * perplexity)  # openTSNE guidance :contentReference[oaicite:5]{index=5}

neighbors, distances = knn_from_annoy(idx, k=k)

# Wrap precomputed neighbor graph for openTSNE
knn_index = PrecomputedNeighbors(neighbors=neighbors, distances=distances)

# Build affinities from your precomputed kNN
affinities_trai = PerplexityBasedNN(data=None, perplexity=perplexity, knn_index=knn_index)

# Fit t-SNE using the precomputed affinities
tsne = TSNE(
    n_components=2,
    initialization="spectral",  # safe when youâ€™re not relying on X for PCA init
    random_state=0,
    n_jobs=8,
)
embedding_train = tsne.fit(X=None, affinities=affinities_trai)
```
